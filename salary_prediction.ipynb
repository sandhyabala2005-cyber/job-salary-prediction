```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:/Users/HP/Downloads/python_dataset.csv")
df.head()
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
def salary_category(s):
    if s < 30000:
        return "Low"
    elif s < 70000:
        return "Medium"
    else:
        return "High"

df['salary_category'] = df['salary'].apply(salary_category)
```

```python
df = df.drop(['salary'], axis=1)
```

```python
df = df.dropna()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

cols = ['job_title', 'education_level', 'industry', 'location']

for col in cols:
    df[col] = le.fit_transform(df[col])

# target encoding
df['salary_category'] = le.fit_transform(df['salary_category'])
```

```python
X = df.drop('salary_category', axis=1)
y = df['salary_category']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)
```

```python
X = pd.get_dummies(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

dt = DecisionTreeClassifier()
rf = RandomForestClassifier()

dt.fit(X_train, y_train)
rf.fit(X_train, y_train)

print("DT Accuracy:", dt.score(X_test, y_test))
print("RF Accuracy:", rf.score(X_test, y_test))
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
